```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sb
from sklearn.metrics import (accuracy_score)
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

scaling



```python
x_train=x_train/255
x_test=x_test/255
```

```python
print(x_train[0])
```

image with size28*28
flatten-from n-dimensional to 1-dim
sigmoid-gives probability for every class label

```python
model=keras.Sequential([keras.layers.Flatten(input_shape=(28,28)),keras.layers.Dense(10,activation='sigmoid')]) #creation of network
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])  #compilation
```

```python
#training models
model.fit(x_train,y_train,epochs=5)
```

```python
model.evaluate(x_test,y_test)
```
